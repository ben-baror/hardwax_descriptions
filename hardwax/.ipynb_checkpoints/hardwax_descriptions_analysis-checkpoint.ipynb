### Importing, loading, cleaning

```python
import pandas as pd
import numpy as np

from textacy import preprocessing, text_stats, extract
import spacy
from spacy.tokens import Doc

#Wordcloud
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
from pypalettes import load_cmap

nlp = spacy.load("en_core_web_md") 
pd.set_option('display.max_colwidth', None)
```

```python
# Load and save Dataframe, exclude NaN values

df = pd.read_csv('descriptions2.csv',
           usecols=['descriptions'])

print(f'NaN values:\n{df.isna().sum()}')

df = df.dropna()

df.tail(3)
```

```python
# basic cleaning function, each description gets lowered and punctioation is removed
def lower_replace(series):
    clean_series = (series.str.lower()
           .str.replace(r'\[.*?\]','', regex=True)
           .str.replace('-', ' ', regex=False)
           .str.replace(r'[^\w\s]','', regex=True)
           .str.replace('\xa0', ' ', regex=False)
           .str.replace('\r\n', ' ', regex=False)
          )

    return clean_series
```

```python
### create column:  lowered descrpitions, punctuation removed 
df['lower_no_punc'] = lower_replace(df['descriptions'])

norm_white_space = preprocessing.make_pipeline(
    preprocessing.normalize.whitespace)

# Normalise whitespace
df['lower_no_punc'] = df['lower_no_punc'].apply(norm_white_space)
```

```python
### create column: lemmatised desctiptions ###

descriptions = df['lower_no_punc'].tolist()     # save the lower_no_punc into a list
lemmas = []                                     # empty container where we'll store the lemmas

# let spacy process texts in batches instead of one by one
for one_sescrpition in nlp.pipe(descriptions, batch_size=100, disable=["ner", "parser"]):
    lemmas.append(' '.join([token.lemma_ for token in one_sescrpition if not token.is_stop]))

df['lemmatized'] = lemmas
```

```python
### create column: ENTITIES ##################

raw_descriptions = df['descriptions'].tolist()        # save the (raw) descriptions column into a list
ents = []                                             # empty container where we'll store the entities

for one_description in nlp.pipe(raw_descriptions, batch_size=100):
    ents.append([ent.text for ent in one_description.ents])

df['entitised'] = ents
```

```python
### create column: VERBS #####################

raw_descriptions = df['descriptions'].tolist()        # save the (raw) descriptions column into a list
verbs = []                                             # empty container where we'll store the verbs

for one_description in nlp.pipe(raw_descriptions, batch_size=100, disable=["ner","parser"]):
    verbs.append([word.lemma_ for word in one_description if word.pos_ == 'VERB'])

df['VERBS'] = verbs
```

### Analysis

#### Statistics with Textacy

```python
d = df['descriptions']+'.' # add '.' at the end of each description
```

Combining all descriptions creates a very large document so need break it up into a into smaller units and then we combine the doc objects into a new single doc object:

```python
# list of doc objects
docs = list(nlp.pipe(d.to_list()))
# combine all docs into one object
c_doc = Doc.from_docs(docs)
```

```python
# Number of words
text_stats.basics.n_words(c_doc)
```

```python
# Number of unique words
text_stats.basics.n_unique_words(c_doc)
```

```python
# Lexical Density: 
# A direct ratio of the number of unique words (types) to all words (tokens)
print(f'{round(text_stats.diversity.ttr(c_doc)*100,2)} of the words are unique')
```

A TTR of 3.81% means the text is highly repetitive. A very small vocabulary is used repeatedly.

##### More fine grained stats:

```python
# Count the number of times rach syntactic dependency relation appears as a token annotation
text_stats.counts.dep(c_doc)
```

```python
# Count the number of times each coarsed-grained universal part-of-speech tag appears as a token annotation 
text_stats.counts.pos(c_doc)
```

```python
# Count the number of times each fine-grained part-of-speech tag appears as a token annotation 
text_stats.counts.tag(c_doc)
```

```python
# Countr the number of times each value for a morphological feature appears as a token annotation
text_stats.counts.morph(c_doc)
```

```python
# Readability measure from 0 (most difficult) to 100 (easiest)
text_stats.readability.flesch_reading_ease(c_doc)
```
